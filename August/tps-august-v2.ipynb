```python
train_df = pd.read_csv("/kaggle/input/tabular-playground-series-aug-2021/train.csv")
test_df = pd.read_csv("/kaggle/input/tabular-playground-series-aug-2021/test.csv")
```

```python
target=train_df.loss
train_df = train_df.drop("loss", axis = 1)
```

```python
from sklearn.feature_selection import mutual_info_regression

def mutual_info(X,y):
    mi_scores = mutual_info_regression(X,y) # gives the information about the mutual info betwee two variabls, if they are dep or independent
    mi_scores = pd.Series(mi_scores, name="Mi_scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores
```

```python
target
```

```python
train_df
```

```python
corr = mutual_info(train_df[:10000], target[:10000]) # correlated features
# pd.set_option('display.max_rows', None)
```

```python
feats = corr[corr > 0] # best features 
```

```python
X = train_df[feats.index] # train data
y = target # final target data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 123)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

model = XGBRegressor(n_estimators=100, learning_rate=0.01)
model.fit(X_train, y_train, early_stopping_rounds=10, eval_set=[(X_test, y_test)])
```

```python
from sklearn.metrics import mean_absolute_error

valid_predictions = model.predict(X_test)
print("Mean Absolute Error: " + str(mean_absolute_error(valid_predictions, y_test)))
```

```python
res = model.predict(test_df[feats.index])
```

```python
res = res.reshape(-1,1)
res = [round(val[0]) for val in res]
```

```python
sub = pd.read_csv("/kaggle/input/tabular-playground-series-aug-2021/sample_submission.csv")
```

```python
sub.loss = res
```
